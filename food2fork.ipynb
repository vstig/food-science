```python
from urllib2 import Request, urlopen, URLError
import json
```

```python
api_key = '7a1ad04206b45bea1698a3ef6237b79f'
```

```python
def search_recipes(search_term):
    request = Request('http://food2fork.com/api/search?key={}&q={}'.format(api_key, search_term))
    try:
        response = urlopen(request)
        recipes = response.read()
        return [r['recipe_id'] for r in json.loads(recipes)['recipes']]
    except URLError, e:
        print "No recipes. Got an error code:",e
        return None
    
def get_recipe(recipe_id):
    request = Request('http://food2fork.com/api/get?key={}&rId={}'.format(api_key, recipe_id))
    try:
        response = urlopen(request)
        recipe = response.read()
        return json.loads(recipe)
    except URLError, e:
        print "No recipe found for id. Got an error code:",e
        
def get_recipes(search_list, fname='recipe_data.dat'):
    for search_term in search_list:
        rIds = search_recipes(search_term)
        with open(fname, 'a') as f:
            for rId in rIds:
                recipe = get_recipe(rId)
                recipe['search_term'] = search_term
                f.write('{}\n'.format(json.dumps(recipe)))
```

```python
search_list = ['soup', 'breakfast', 'dinner', 'pasta', 'potatoes', 'beans', 'salad', 'onion', 'basil', 'ginger']

get_recipes(search_list)
```
